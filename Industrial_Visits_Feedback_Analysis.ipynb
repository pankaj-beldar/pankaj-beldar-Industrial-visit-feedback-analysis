```python
import pandas as pd
df= pd.read_excel('/content/drive/MyDrive/Industrial visit/Industrial Visit Feedback form (Responses) (2).xlsx')
df.head()
```

```python



# List of columns from Q1 to Q4
question_columns = ['Q1', 'Q2', 'Q3', 'Q4']

# Calculating total responses for each question
total_responses = df[question_columns].count()

# Calculating percentage of each response option for each question
for question_column in question_columns:
    strongly_agree_count = df[df[question_column].str.contains(r'Strongly Agree \(100- 80%\)', regex=True, na=False)][question_column].count()
    moderately_agree_count = df[df[question_column].str.contains(r'Moderately Agree \(80- 60%\)', regex=True, na=False)][question_column].count()
    neutral_count = df[df[question_column].str.contains(r'Neutral \(40- 20%\)', regex=True, na=False)][question_column].count()
    agree_count = df[df[question_column].str.contains(r'Agree \(60- 40%\)', regex=True, na=False)][question_column].count()

    total_responses_for_question = total_responses[question_column]

    print(f"Question-wise percentage for {question_column}:")
    print("Strongly Agree (100- 80%):", (strongly_agree_count / total_responses_for_question) * 100, "%")
    print("Moderately Agree (80- 60%):", (moderately_agree_count / total_responses_for_question) * 100, "%")
    print("Agree (60- 40%):", (agree_count / total_responses_for_question) * 100, "%")
    print("Neutral (40- 20%):", (neutral_count / total_responses_for_question) * 100, "%")

    print()


```

```python
import pandas as pd

# Assuming df contains your data
# Create a DataFrame containing only the responses (Q1 to Q4)
responses_df = df[['Name of the Industry Visited','Year', 'Q1', 'Q2', 'Q3', 'Q4']]

# Define a function to calculate the percentage of 'Moderately Agree (80-60%)' and 'Strongly Agree (100-80%)' responses
def calculate_percentage(response_column):
    return ((response_column.str.contains('Moderately Agree') |
             response_column.str.contains('Strongly Agree')).sum() / len(response_column)) * 100

# Group by industry and calculate the percentage of 'Moderately Agree' and 'Strongly Agree' responses for each question
industry_response_percentages = responses_df.groupby(['Name of the Industry Visited','Year']).agg({
    'Q1': calculate_percentage,
    'Q2': calculate_percentage,
    'Q3': calculate_percentage,
    'Q4': calculate_percentage
}).reset_index()

# Calculate the average of all four questions for each visit and add a new column 'Average'
industry_response_percentages['Average_feedback'] = industry_response_percentages[['Q1', 'Q2', 'Q3', 'Q4']].mean(axis=1)

# Add a new column 'Attainment' based on average values
industry_response_percentages['Attainment_Level'] = industry_response_percentages['Average_feedback'].apply(lambda x: 3 if x > 70 else (2 if x > 60 else 1))

# Print or display the industry_response_percentages DataFrame
industry_response_percentages






```

# **Sentiment Analysis**

```python
import pandas as pd
df= pd.read_excel('/content/drive/MyDrive/Industrial visit/Industrial Visit Feedback form (Responses) (2).xlsx')
df.head()
```

```python
import string
from textblob import TextBlob

# Now you can use TextBlob functionalities

def clean_text(text):
    if isinstance(text, str):
        # Remove punctuation and non-alphabetic characters
        text = ' '.join(word for word in text.split() if word.isalpha())
        return text.translate(str.maketrans('', '', string.punctuation))
    else:
        # If the data is not a string, return an empty string
        return ''

# Apply text cleaning to the 'comments' column
df['Cleaned Comments'] = df['comments'].apply(clean_text)

# Remove rows that are empty after cleaning or have no text
df = df[df['Cleaned Comments'].apply(lambda x: len(x.strip()) > 0)]

# Define a function to get sentiment polarity using TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Apply sentiment analysis to the cleaned text in 'Cleaned Comments' column
df['Sentiment'] = df['Cleaned Comments'].apply(get_sentiment)

# Display the DataFrame with sentiment analysis results
print(df[['comments', 'Cleaned Comments', 'Sentiment']])
```

```python
from textblob import TextBlob

# Assuming 'df' is your DataFrame and 'Any Other Suggestion' column is named 'Any Other Suggessation'
# Fill any missing values with an empty string
df['Cleaned Comments'].fillna('', inplace=True)

# Define a function to categorize sentiment based on polarity score
def categorize_sentiment(score):
    if score > 0:  # Positive sentiment
        return 'Positive'
    elif score < 0:  # Negative sentiment
        return 'Negative'
    else:  # Neutral sentiment
        return 'Neutral'

# Define a function to get sentiment polarity using TextBlob and categorize sentiments
def get_sentiment_category(text):
    analysis = TextBlob(text)
    sentiment_score = analysis.sentiment.polarity
    return categorize_sentiment(sentiment_score)

# Apply sentiment analysis and categorization to the 'Any Other Suggessation' column
df['Sentiment Category'] = df['Cleaned Comments'].apply(get_sentiment_category)

# Display the DataFrame with categorized sentiment analysis results
df[['Cleaned Comments', 'Sentiment Category']]
```

# **Binary seniment**

```python
# Function to clean text and handle non-string data
def clean_text(text):
    if isinstance(text, str):
        text = ' '.join(word for word in text.split() if word.isalpha())
        return text.translate(str.maketrans('', '', string.punctuation))
    else:
        return ''

# Apply text cleaning to the 'comments' column
df['Cleaned Comments'] = df['comments'].apply(clean_text)

# Function to categorize sentiment into positive and negative
def categorize_sentiment(score):
    if score >= 0:
        return 'Positive'
    else:
        return 'Negative'

# Function to get sentiment polarity using TextBlob and categorize sentiments
def get_sentiment_category(text):
    analysis = TextBlob(text)
    sentiment_score = analysis.sentiment.polarity
    return categorize_sentiment(sentiment_score)

# Apply sentiment analysis and categorization to the 'Cleaned Comments' column
df['Sentiment'] = df['Cleaned Comments'].apply(get_sentiment_category)

# Count occurrences of positive and negative sentiments along with comments
positive_comments = df[df['Sentiment'] == 'Positive']['Cleaned Comments']
negative_comments = df[df['Sentiment'] == 'Negative']['Cleaned Comments']

# Display counts and comments for both sentiments
print(f"Positive Sentiments Count: {len(positive_comments)}")
print(f"Negative Sentiments Count: {len(negative_comments)}")
```

```python
import matplotlib.pyplot as plt

# Positive and Negative sentiment counts
positive_count = len(positive_comments)
negative_count = len(negative_comments)

# Data to plot
sentiment_counts = [positive_count, negative_count]
labels = ['Positive Sentiments', 'Negative Sentiments']
colors = ['#ff9999', '#66b3ff']  # Colors for each section

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sentiment_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Sentiments')

# Display the pie chart
plt.show()
```
